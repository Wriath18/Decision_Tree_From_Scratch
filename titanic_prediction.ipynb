```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python

trainning_data = pd.read_csv('train.csv')
testing_data = pd.read_csv('test.csv')
```

```python
print(trainning_data.shape)
```

```python
print(testing_data.shape)
```

```python
trainning_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
trainning_data['Sex'] = le.fit_transform(trainning_data['Sex'])
```

```python
trainning_data.head()
```

```python
testing_data['Sex'] = le.fit_transform(testing_data['Sex'])
```

```python
testing_data.head(
)
```

```python
necessary_data = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']
output_cols = ['Survived']
```

```python
sktree = DecisionTreeClassifier(criterion='entropy')
```

```python
sktree.fit(trainning_data[necessary_data], trainning_data[output_cols])
```

```python
y_predictions = sktree.predict(testing_data[necessary_data])
```

```python
y_predictions
```

```python
passid = testing_data['PassengerId'].values
```

```python
passid
```

```python
y_predictions
```

```python
df = pd.DataFrame(
    {
        'PassengerId':passid,
        'Survived' : y_predictions
    }
)

df.to_csv('predictions2.csv', index=False)
```
